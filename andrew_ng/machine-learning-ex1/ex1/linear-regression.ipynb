```python
x = data[:, :1]
x, x.shape
```

```python
y = data[:, 1:]
y, y.shape
```

```python
plt.plot(x, y, 'rx')
plt.ylabel('Profit in $10,000s')
plt.xlabel('Population of City in 10,000s')
plt.show()
```

```python
row, col = data.shape
X = np.append(np.ones((row, 1)), x, axis=1)
X
```

```python
theta = np.zeros((2,1))
iterations = 1500
alpha = 0.01
theta, iterations, alpha
```

```python
J = 0
```

```python
h = np.dot(X, theta)
```

```python
def computeCost(X, y, theta):
    m, col = y.shape
    h = np.dot(X, theta)
    J = (1 / (2*m)) * np.sum(np.power((h-y), 2))
    return J

computeCost(X, y, theta)
```

```python
def gradientDescent(X, y, theta, alpha, num_iters):
    m, col = y.shape
    J_history = np.zeros(num_iters)
    for iter in np.arange(num_iters):
        h = np.dot(X, theta)
        theta = theta - np.dot(X.T, (alpha / m) * (h - y))

        # Save the cost J in every iteration    
        J_history[iter] = computeCost(X, y, theta)
    return theta
        
theta_final = gradientDescent(X, y, theta, alpha, iterations)
theta_final
```

```python
plt.plot(x, y, 'rx')
plt.plot(x, np.dot(X, theta_final), '-')
plt.ylabel("Linear regression")
plt.xlabel("Training data")
plt.show()
```

```python
predict_1 = np.dot([1, 3.5], theta_final)
predict_1
```
